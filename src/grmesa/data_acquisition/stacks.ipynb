```python
import py_gamma as pg

import numpy as np
import pandas as pd
import xarray as xr
import rioxarray as rxa

import matplotlib.pyplot as plt

from uavsar_pytools.convert.tiff_conversion import read_annotation, grd_tiff_convert

from pathlib import Path
```

```python
stack_dir = Path('/data/coherence/stack')
out_dir = Path('/data/coherence/gamma')
```

```python
for ann_fp in stack_dir.glob('*.ann'):
    dop = list(stack_dir.glob(f'*.dop'))
    assert len(dop) == 1
    dop = dop[0]
    ann = read_annotation(ann_fp)
    date = ann['start time of acquisition']['value']

    for track in range(1,2):
        fps = list(stack_dir.glob(f'{ann_fp.stem}_s{track}*.slc'))
        assert len(fps) == 1
        fp = fps[0]

        pg.par_UAVSAR_SLC(ann_fp, fp, out_dir.joinpath(date.strftime('%Y%m%d.slc.par')), out_dir.joinpath(date.strftime('%Y%m%d.slc')), '-', '-', dop)
```
