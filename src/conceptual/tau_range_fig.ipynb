```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
days = np.array(range(50))
```

```python
def temporal_decay(day, gamma_inf, tau):
    coherence = (1 - gamma_inf) * np.exp(-day/tau) + gamma_inf
    return coherence
temporal_decay = np.vectorize(temporal_decay)
```

```python
fig, axes = plt.subplots(2,1, figsize = (12,8))
for tau in [5, 10, 15]:
    axes[0].plot(temporal_decay(days, 0.3, tau), label = f'tau: {tau} days')
for gamma_inf in [0.2, 0.3, 0.4]:
    tau = 10
    axes[1].plot(temporal_decay(days, gamma_inf, tau), label = f'Gamma Infinity: {gamma_inf}')

axes[0].set_title(f'Gamma Infinity: {0.3}')
axes[1].set_title(f'Tau: {10} days')

for ax in axes:
    ax.legend()
    ax.set_xlabel('Days between images')
    ax.set_ylabel('Coherence')
plt.tight_layout()

figdir = Path('../../figures/conceptual')
plt.savefig(figdir.joinpath('temporal_decay_range.png'))
```
