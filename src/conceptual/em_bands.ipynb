```python
import numpy as np
import pylab as plt

def rect(x,y,w,h,c, a):
    ax = plt.gca()
    polygon = plt.Rectangle((x,y),w,h,color=c, alpha = a)
    ax.add_patch(polygon)

def rainbow_fill(ax, X,Y, cmap=plt.get_cmap("jet_r")):
    ax.plot(X,Y,lw=0)  # Plot so the axes scale correctly

    dx = X[1]-X[0]
    N  = float(X.size)

    alpha_n = 50
    for n, (x,y) in enumerate(zip(X,Y)):
        color = cmap(n/N)
        if n < alpha_n: alpha = n / alpha_n
        elif len(X) - n < alpha_n: alpha = (len(X) - n) / alpha_n
        else: alpha = 1
        rect(x,-1.1,dx,y*2.2,color, alpha)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize = (12, 2))
ax.set_facecolor('black')
x = np.arange(0, 100, 0.001)
def f(x): return 1.03**x
y = -np.sin(2*np.pi*f(x)*x*0.01)

# Test data    
X = np.linspace(55,65,300)
Y = np.ones_like(X)
rainbow_fill(ax, X,Y)
# plt.show()

plt.plot(x, y, color = 'white')
plt.xticks([])
plt.yticks([])

plt.savefig(fig_dir.joinpath('full_em_rainbow.png'),bbox_inches='tight')
```

```python
import numpy as np
import pylab as plt

def rect(x,y,w,h,c):
    ax = plt.gca()
    polygon = plt.Rectangle((x,y),w,h,color=c)
    ax.add_patch(polygon)

def rainbow_fill(X,Y, cmap=plt.get_cmap("jet")):
    plt.plot(X,Y,lw=0)  # Plot so the axes scale correctly

    dx = X[1]-X[0]
    N  = float(X.size)

    for n, (x,y) in enumerate(zip(X,Y)):
        color = cmap(n/N)
        rect(x,0,dx,y,color)

# Test data    
X = np.linspace(0,10,100)
Y = .25*X**2 - X
rainbow_fill(X,Y)
plt.show()
```
