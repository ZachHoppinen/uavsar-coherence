```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

fig, axes = plt.subplots(2, 1, figsize = (8,8))

# Meshgrid
x, y = np.meshgrid(np.linspace(-5, 5, 10), 
                   np.linspace(-5, 5, 10))
  
ax = axes[0]
# Directional vectors
u = 2* np.ones_like(x) + np.random.random((len(x), len(y)))/1.5
v = 2 - np.random.random((len(x), len(y)))/1.5
  
# Plotting Vector Field with QUIVER
ax.quiver(x, y, u, v, color='g')
ax.set_title('High Coherence Phases')

ax = axes[1]
# Directional vectors
u = 2* np.ones_like(x) + np.random.random((len(x), len(y))) * 2.2
v = 2 - np.random.random((len(x), len(y))) * 2.2
  
# Plotting Vector Field with QUIVER
ax.quiver(x, y, u, v, color='g')
ax.set_title('Low Coherence Phases')

for ax in axes:
    # Setting x, y boundary limits
    ax.set_xlim(-7, 7)
    ax.set_ylim(-7, 7)

    ax.tick_params(left = False, right = False , labelleft = False ,
                    labelbottom = False, bottom = False)
    
    # Show plot with grid
    # ax.grid()
figdir = Path('../../figures/conceptual')
plt.savefig(figdir.joinpath('coherence_quivers.png'))
```

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

fig, axes = plt.subplots(2, 1, figsize = (8,8))

# Meshgrid
x, y = np.meshgrid(np.linspace(-5, 5, 10), 
                   np.linspace(-5, 5, 10))
  
ax = axes[0]
# Directional vectors
u = 2* np.ones_like(x) + np.random.random((len(x), len(y)))/5
v = 2 - np.random.random((len(x), len(y)))/5
  
# Plotting Vector Field with QUIVER
ax.quiver(x, y, u, v, color='g')
ax.set_title('High Amplitude Return')

ax = axes[1]
# Directional vectors
u = 0.2* np.ones_like(x) + np.random.random((len(x), len(y))) /5
v = 0.2 - np.random.random((len(x), len(y)))/5
  
# Plotting Vector Field with QUIVER
ax.quiver(x, y, u, v, color='g', angles='xy', scale_units='xy', scale=1.)
ax.set_title('Low Amplitude Return')

for ax in axes:
    # Setting x, y boundary limits
    ax.set_xlim(-7, 7)
    ax.set_ylim(-7, 7)

    ax.tick_params(left = False, right = False , labelleft = False ,
                    labelbottom = False, bottom = False)
    
    # Show plot with grid
    # ax.grid()
figdir = Path('../../figures/conceptual')
plt.savefig(figdir.joinpath('amplitude_quivers.png'))
```
